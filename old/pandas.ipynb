```python
import pandas as pd
reader = pd.read_csv('ctauto.csv', iterator=True)
try:
    df = reader.get_chunk(100000000)
except StopIteration:
    print("Iteration is stopped.")
```

```python
loop = True
chunkSize = 100000
chunks = []
while loop:
    try:
        chunk = reader.get_chunk(chunkSize)
        chunks.append(chunk)
    except StopIteration:
        loop = False
        print("Iteration is stopped.")
df = pd.concat(chunks, ignore_index=True)
```
